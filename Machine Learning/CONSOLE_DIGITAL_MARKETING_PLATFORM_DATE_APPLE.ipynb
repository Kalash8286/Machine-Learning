```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
```

```python
datasetDate = pd.read_csv("CONSOLE_DIGITAL_MARKETING_PLATFORM_DATE_APPLE.csv", sep=",")
dataset = datasetDate.loc[:, datasetDate.columns != 'DATE']
```

```python
print(datasetDate.columns)
print(dataset.columns)
```

```python
print(dataset.shape)
```

```python
dataset.head(5)
```

```python
print(dataset.describe())
```

```python
################## HISTOGRAM  ######################
dataset.hist(figsize=(30,25))
plt.show()
```

```python
dataset.plot(figsize=(30,25))
```

```python
dataset.plot.box(figsize=(30,25))
```

```python
################# SUBPLOT GRAPH #####################

axs = dataset.plot.area(figsize=(30,25), subplots=True)
```

```python
################# BAR GRAPH #####################

plt.figure(figsize=(150,50))

for (columnName, columnData) in dataset.iteritems():
  plt.bar(columnData.values, datasetDate.DATE, label=columnName,width=.5)

  
plt.legend()
plt.show()
```

```python
################# SCATTER GRAPH #####################

plt.figure(figsize=(50,50))

for (columnName, columnData) in dataset.iteritems():
  plt.scatter(columnData.values, datasetDate.DATE, label=columnName )
  

plt.legend()
plt.show()
```
