```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
import seaborn as sns
from pandas.plotting import scatter_matrix
import matplotlib.pyplot as plt
```

```python
################# BAR GRAPH #####################

# tutorials = ['C_COMPILER', 'C_TUTORIAL', 'CSHARP_COMPILER', 'CSHARP_TUTORIAL', 'CPLUSPLUS_COMPILER', 'CPLUSPLUS_TUTORIAL', 'CALCULATORPLUS', 'CALENDARPLUS', 'COMPILERPLUS', 'CSS_TUTORIAL', 'DART_COMPILER', 'DART_TUTORIAL', 'GO_COMPILER', 'GO_TUTORIAL',
#        'HTML_TUTORIAL', 'JAVA_COMPILER', 'JAVA_TUTORIAL', 'JAVASCRIPT_COMPILER', 'JAVASCRIPT_TUTORIAL', 'KOTLIN_COMPILER', 'KOTLIN_TUTORIAL', 'MOTIVATIONPLUS', 'NOTESPLUS', 'PHP_COMPILER', 'PHP_TUTORIAL', 'PYTHON_COMPILER', 'PYTHON_TUTORIAL', 'RUBY_COMPILER',
#        'RUBY_TUTORIAL', 'SUCCESS_STORIES', 'SWIFT_COMPILER', 'SWIFT_TUTORIAL', 'TASKSPLUS', 'TIMEPLUS', 'TRAVEL_DIARIES', 'XML_TUTORIAL']

colorIndex = 0
plt.figure(figsize=(150,50))

for (columnName, columnData) in dataset.iteritems():
  plt.bar(columnData.values, datasetDate.DATE, color=colors[colorIndex],label=columnName,width=.5)
  colorIndex += 1
  
plt.legend()
plt.show()
```

```python
################# SCATTER GRAPH #####################

plt.figure(figsize=(50,50))

colorIndex = 0
for (columnName, columnData) in dataset.iteritems():
  plt.scatter(columnData.values, datasetDate.DATE, label=columnName, color=colors[colorIndex])
  colorIndex += 1

# plt.scatter(datasetDate.DATE, dataset.C_COMPILER, label="C_COMPILER",color="c")
# plt.scatter(datasetDate.DATE, dataset.C_TUTORIAL, label="C_TUTORIAL",color="g")
plt.legend()
plt.show()
```
